```python
import pyspark.sql.functions as F
from pyspark.sql.types import StructType, StructField, StringType, IntegerType, DoubleType, DateType, TimestampType
```

```python
catalog = "nyc_yellow_taxi"
data_path = "/Volumes/nyc_yellow_taxi/source_data/raw/2025_data/*.parquet"
```

```python
df = spark.read.option("inferSchema", "true").option("header", "true").parquet(data_path)
df = df.withColumn("ingestion_time", F.current_timestamp()) \
        .withColumn("_source_file", F.col("_metadata.file_path"))

display(df.limit(10))
```

```python
df.write.mode("overwrite").format("delta").option("mergeSchema", "true").saveAsTable(f"{catalog}.bronze.trips")
```

### Zone Lookup

```python
file_path = "/Volumes/nyc_yellow_taxi/source_data/raw/taxi_zone_lookup/taxi_zone_lookup.csv"

df = spark.read.option("inferSchema", "true").option("header", "true").csv(file_path)
df = df.withColumn("ingestion_time", F.current_timestamp()) \
        .withColumn("_source_file", F.col("_metadata.file_path"))

display(df.limit(10))
```
